## 1 Assume that $V$ is constant in time and equal to $V_{max}$ , and that moving left or right costs $a$ unit of energy. Model the problem as a MDP, and explain how to compute the optimal solution for a given discount factor $\gamma$.

## MDP problem

States $S$: the combination of orientation of turbine $s_t$ and the orientation of wind $s_w$. We use a tuple $s = (s_t,s_w)$ the represent a state.

Actions $A$: is the control the angle of the turbine, we just have three action in a given state: one degree left, staty still or one degree right.

$A = \{-1, 0, 1\}$, for un action $a$, $a\in A$

$Reward(s, a, s')$ : reward for the transition $(s, a, s')$. In this case :

\begin{equation}
\label{eq6}
Reward(s, a, s')=\left\{
\begin{aligned}
f(V)cos(\theta)& , & a = 0, \\
f(V)cos(\theta) - a & , & otherwise.
\end{aligned}
\right.
\end{equation}0

As indicated in the project, $\theta$ is the difference of angle between the wind orientation and the turbine orientation, and $V$ the wind speed

We can use value iteration [Bellman, 1957]  to compute the optimal solution. Assume $V_{opt}$ is the optimal value

(1) Initialize $V_{opt}^{0}(s) \leftarrow 0$

(2) For each iteration $t = 1, ...t_{VI}:$

(3) For each state $s$:

(4) $V_{opt}^{t}(s) \leftarrow \underset{a \in A(s)}{\operatorname{\max}} Reward(s, a, s') + \gamma V_{opt}^{t-1}(s')$

(5) If the norm of $V^t - V^{t-1} < \epsilon$, break; else continue to iterate.

## 2 Implement an algorithm that finds the optimal solution and illustrate the optimal solution.

Assume at first, the orientation of turbine is equal to the orientation of wind.

The length of the sequence of wind is $N$, the length of the sequence of turbine is $M = 2*N - 1$

For example, if $N = 3$ and the sequence of wind is [90, 91, 92], since at each state we just have 3 actions, so the sequence of turbine is [88, 89, 90, 91, 92]

```python
# The length of the sequence of wind is N
N = 7
# The length of the sequence of turbine is M
M = 2*N - 1
epsilon = 1e-15
```

```python
# Generate Wind data
# Assume the initial orientation is 90 degree
wind_init = 90
Wind = [wind_init]
# The wind orientation moves by plus or minus one degree at each time step
orientation_change = np.array([-1, 1])
for i in range(N-1):
    Wind.append(Wind[-1]+orientation_change[np.random.randint(0, 2)])
print(Wind)
```

```python
# Assume the original orientation is equal to Wind[0]
Turbine = np.array([i for i in range(Wind[0]-N+1, Wind[0]+N)])
print(Turbine)
```

```python
# Create a matrix with the size 2*M*N since we just need to compare the last two steps of optimal values
V_opt = np.zeros([2, M, N])
```

```python
# Compute the theta which means the absolute value of the difference between the orientation of wind and that of turbine
def theta(angle_w, angle_t):
    return np.abs(angle_w - angle_t)*np.pi/180
```

```python
# From a given position, compute the values with the possible actions
def find_v(i, j, V_opt, a, gamma, f_v, Wind):
    return np.array([f_v[j+1]*np.cos(theta(Wind[j+1], Turbine[i-1])) - a + gamma*V_opt[0, i-1, j+1], \
                   f_v[j+1]*np.cos(theta(Wind[j+1], Turbine[i])) + gamma*V_opt[0, i, j+1],\
                    f_v[j+1]*np.cos(theta(Wind[j+1], Turbine[i+1])) - a + gamma*V_opt[0, i+1, j+1]])
```

```python
# Compute the optimal value and return the optimal policy
def compute_optimal(a, gamma, Wind, f_v=np.ones(N)):
    while True:
        # Create a optimal policy matrix with initial value 10, we use 10 just to distingush -1, 0, 1, we use 10 means it is not accessible. 
        pi_opt = np.ones([M, N],dtype=int)*10
        for j in range(N-1):
            for i in range(N-1-j, N+j):
                tmp = find_v(i, j, V_opt, a, gamma, f_v, Wind)
                V_opt[1, i, j] = np.max(tmp)
                pi_opt[i, j] = np.argmax(tmp) - 1
        #print(V_opt[1])
        if np.linalg.norm(V_opt[0]-V_opt[1]) < epsilon:
            break    
        V_opt[0] = V_opt[1]
    return pi_opt

# According to a policy, find the path and return the sequence of actions
def find_policy(pi_opt):
    policy = []
    init = N-1
    for i in range(N-1):
        case = pi_opt[init, i]
        policy.append(case)
        init = init + case
    return policy
```

```python
def optimal_policy(a, gamma):
    pi_opt = compute_optimal(a, gamma, Wind)
    policy = find_policy(pi_opt)
    print(policy)
```

```python
# Assume a = 0, gamma = 0, that is a greedy algorithm
optimal_policy(0, 0)
```

```python
# Assume a = 1, gamma = 0
optimal_policy(1, 0)
```

We can notice that if a is very large, that means we need a greate amount of energy to turn the turbine to left or right, so the best solution is to stay still.

```python
# Assume a = 0, gamma = 1
optimal_policy(0, 1)
```

```python
a = np.linspace(0, 1, 5)
gamma = np.linspace(0, 1, 5)
```

```python
for a_i in a:
    for gamma_j in gamma:
        print(a_i, gamma_j)
        optimal_policy(a_i, gamma_j)
```

## 3. Assume now that the wind at time $t + 1$ evolves as $V (t + 1) = max(0, min(1.1, V (t) + b(t)))$

```python
print(np.min([1.0, 2.0]))
```

```python
# 2.1.2

def v_t_plus_one(v_t, b):
    delta_t = np.random.uniform(low=-1.0, high=1.0)
    return np.max([0, np.min([1.1, v_t + b*delta_t])])

def create_v(b):    
    v = [0]
    for i in range(1, N):
        v.append(v_t_plus_one(v[-1], b))
    return v

def calculate_f_v(b):
    tmp = create_v(b)
    return np.power(tmp, 3)
```

```python
def optimal_policy_evolve(a, gamma, b):
    f_v = calculate_f_v(b)
    pi_opt = compute_optimal(a, gamma, f_v)
    policy = find_policy(pi_opt)
    print(policy)
```

```python
a = np.linspace(0, 1, 2)
gamma = np.linspace(0, 1, 2)
b = np.linspace(0.01, 5, 5)
for a_i in a:
    for gamma_j in gamma:
        for b_k in b:
            print(a_i, gamma_j, b_k)
            optimal_policy_evolve(a_i, gamma_j, b_k)
```
