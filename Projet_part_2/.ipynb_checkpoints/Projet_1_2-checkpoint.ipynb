```python
import numpy as np
import torch as T
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex
from IPython import display
from math_utils import wrap_to_m180_p180
from wind_turbine import Wind_turbine, Wind
from simu import Random_agent, Basic_agent
from tqdm import tqdm
from collections import deque
from torch.distributions import Categorical
# Code source : https://github.com/paulaubin/wind_turbine_env
```

```python
# Global variable
WIND_SPEED = 10
WIND_HEADING = 270
TURBINE_HEADING = 350
MAX_ITER = 40
```

```python
# Add two lines to the original code because we want to save the sequence of wind speed
class Simu:
    power_output_log = [] 
    action_log = []
    rel_wind_heading_log = []
    true_rel_wind_heading_log = [] 
    wd_heading_log = [] 
    step_count = 0

    def __init__(self, agent=None, wind_model=None, wind_turbine_model=None, max_steps=None):
        self.wd = Wind(10, 0, 1, 'OU') if wind_model is None else wind_model
        self.wt = Wind_turbine(0, False) if wind_turbine_model is None else wind_turbine_model
        self.agent = Basic_agent() if agent is None else agent
        self.max_steps = 24*3600 if max_steps is None else max_steps 
        self.power_output_log = self.max_steps * [0]
        self.action_log = self.max_steps * [1]
        self.rel_wind_heading_log = self.max_steps * [0]
        self.true_rel_wind_heading_log = self.max_steps * [0]
        self.wd_heading_log = self.max_steps * [0]
        # Add self.wd_speed_log to save the the sequence of speed
        self.wd_speed_log = self.max_steps * [0]                    # This line is we added

    def step(self):
        self.rel_wind_heading_log[self.step_count] = wrap_to_m180_p180(self.wd.heading - self.wt.heading)
        self.true_rel_wind_heading_log[self.step_count] = wrap_to_m180_p180(self.wd.heading - self.wt.true_heading)
        self.wd_heading_log[self.step_count] = self.wd.heading
        # Get the speed of wind and save it
        self.wd_speed_log[self.step_count] = self.wd._speed         # This line is we added
        self.action_log[self.step_count] = self.agent.policy(self.rel_wind_heading_log[self.step_count])
        self.power_output_log[self.step_count] = self.wt.step(self.wd.speed, self.wd.heading, self.action_log[self.step_count])
        self.wd.step()

    def run_simu(self):
        while self.step_count < self.max_steps:
            self.step()
            self.step_count += 1

    def __str__(self):
        return str(self.__class__) + ": " + str(self.__dict__)
```

```python
# Initialize a wind instance and the Ornstein-Uhlenbeck model for basic policy and random policy
wd_ba = Wind(WIND_SPEED, WIND_HEADING, 60, 'OU')
wd_ra = Wind(WIND_SPEED, WIND_HEADING, 60, 'OU')

# Initialize a wind turbine for basic policy and random policy
wt_ba = Wind_turbine(TURBINE_HEADING, True)
wt_ra = Wind_turbine(TURBINE_HEADING, True)

# Initialize basic agent
ba = Basic_agent()

# Initialize random agent
ra = Random_agent()

# Set up a simulation that runs for 1 day
sm_ba = Simu(ba, wd_ba, wt_ba, int(np.ceil(24*3600 / wd_ba.step_duration)))
sm_ra = Simu(ra, wd_ra, wt_ra, int(np.ceil(24*3600 / wd_ra.step_duration)))

# Run the simulation
sm_ba.run_simu()
sm_ra.run_simu()

# Get the logs from the simulation
power_output_ba = sm_ba.power_output_log             
power_output_ra = sm_ra.power_output_log            

# Compute the cumulate power and use this value to compare the performance of the two policy
# Cumulate power for basic agent
cum_power_ba = np.cumsum(power_output_ba)

# Cumulate power for random agent
cum_power_ra = np.cumsum(power_output_ra)
```

## 2. Implement an agent that uses a tabular Q-learning algorithm and study its performance and its learning speed.

We have designed two different types of states and implemented them. We use $\theta_t$ to represent the difference in angle between the turbine and wind orientations at time $t$, $\theta_t \in [-180, 180]$. $v_t$ represents the speed of the wind at time $t$. 

(1) First type. We use the $\theta_t$ and $t$ to form a state, which means a state $s = (\theta, v)$. At this time, the speed is a discrete value, which means $v_t \in \{0, 0.5, 1, 1.5, ..., 3* initialSpeed\}$.

(2) Second type. We only use the angle as a state, which means $s = \theta$.

```python
# Get policy accordint to the table value of state and action
def get_policy(Q_table, rel_wind_heading, speed):    
    indice_angle = int(rel_wind_heading) + 180
    indice_speed = int(np.round(speed/0.5))
    # If state = (theta, speed), we get action according to angle and speed in a giving Q table
    if (Q_table.shape[0] == 3):
        Q_action = [Q_table[0, indice_angle, indice_speed],\
                    Q_table[1, indice_angle, indice_speed],\
                    Q_table[2, indice_angle, indice_speed]]        
    # If state = theta, we get action according to angle in a giving Q table
    else:
        Q_action = Q_table[indice_angle]
    return np.argmax(Q_action)
```

```python
# Appling Q learning algorithm with state = (theta, speed)
def Q_learning(policy_flag = 0, initial_speed = WIND_SPEED, wind_heading = WIND_HEADING, trubine_heading = TURBINE_HEADING,
               alpha = 0.01, discout_factor = 0.95, epsion = 0.6, max_iter = MAX_ITER):
    
    # Define the max speed is 3*initial_speed
    len_speed = initial_speed*3
    # Cut the max speed with a stpe size 0.5
    speed_vector = np.arange(0, len_speed, 0.5)
    Q_table = np.zeros((3, 360, len(speed_vector)))

    # Initialize a basic agent instance that will give the basic policy
    ba = Basic_agent()

    # Run the simulation
    for i in tqdm(range(max_iter)):
        # Get a sequence of wind with heading and speed
        wd = Wind(initial_speed, wind_heading, 60, 'OU')
        wt = Wind_turbine(trubine_heading, True)
        sm = Simu(ba, wd, wt, int(np.ceil(24*3600 / wd.step_duration)))
        sm.run_simu()
        wd_heading = sm.wd_heading_log
        wd_speed = sm.wd_speed_log
        
        # Traversal the sequence of wind and compute Q value
        for j in range(len(wd_heading)-1):
            rel_angle = wrap_to_m180_p180(wd_heading[j] - wt.heading)
            speed = wd_speed[j]        
            indice_angle = int(rel_angle) + 180
            indice_speed = int(np.round(speed/0.5))
            # We define two different type of policy, optimal and basic with flag
            # If given flag is 0, we use epsilon greedy with optimal policy, which means select the action with the highest Q value
            if (policy_flag == 0):
                this_Q_A = [Q_table[0, indice_angle, indice_speed],\
                            Q_table[1, indice_angle, indice_speed],\
                            Q_table[2, indice_angle, indice_speed]]
                optimal_action = np.argmax(this_Q_A)
            # If given flag is 1, we use epsilon greedy with basic policy, which means narrow the angle
            else:
                optimal_action = ba.policy(rel_angle)
           
            # epslion greedy    
            if(np.random.uniform(low=0.0, high=1.0) < 1 - epsion):
                action = optimal_action
            else:
                action = int(np.random.randint(0, 3))
            
            indice_action = int(action)
            
            # Compute the reward according state and action
            R = wt.step(speed, wd_heading[j], action)
            
            # Compute the next state after taking the action ang get the indices
            indice_next_angle = int(wrap_to_m180_p180(wd_heading[j+1] - (wt.heading+action-1)) + 180)
            indice_next_speed = int(np.round(wd_speed[j+1]/0.5))
            # Get the Q vaule of next state
            next_Q_A = [Q_table[0, indice_next_angle, indice_next_speed],\
                        Q_table[1, indice_next_angle, indice_next_speed],\
                        Q_table[2, indice_next_angle, indice_next_speed]]
            # Compute the Q(S, A) according to the Q-learning policy
            Q_table[indice_action, indice_angle, indice_speed] += alpha * (R + discout_factor*np.max(next_Q_A) - Q_table[indice_action, indice_angle, indice_speed])        
    return Q_table
```

```python
# Appling Q learning algorithm with state = theta
def Q_learning_sans_speed(initial_speed = WIND_SPEED, wind_heading = WIND_HEADING, trubine_heading = TURBINE_HEADING,
               alpha = 0.01, discout_factor = 0.95, epsion = 0.6, max_iter = MAX_ITER):
   
    Q_table = np.zeros((360, 3))

    ba = Basic_agent()
    
    for i in tqdm(range(max_iter)):
        wd = Wind(initial_speed, wind_heading, 60, 'OU')
        wt = Wind_turbine(trubine_heading, True)
        sm = Simu(ba, wd, wt, int(np.ceil(24*3600 / wd.step_duration)))
        sm.run_simu()
        wd_heading = sm.wd_heading_log
        wd_speed = sm.wd_speed_log
        
        # Traversal the sequence of wind and compute Q value
        for j in range(len(wd_heading)-1):
            rel_angle = wrap_to_m180_p180(wd_heading[j] - wt.heading)
            speed = wd_speed[j]        
            indice_angle = int(rel_angle) + 180        
            this_Q_A = Q_table[indice_angle]
            
            # We use the basic policy as the optimal policy, this is the epsilon greedy part
            optimal_action = ba.policy(rel_angle)
            if(np.random.uniform(low=0.0, high=1.0) < 1 - epsion):
                action = optimal_action
            else:
                action = int(np.random.randint(0, 3))
            indice_action = int(action)
            # Compute the reward according state and action
            
            R = wt.step(speed, wd_heading[j], action)
            indice_next_angle = int(wrap_to_m180_p180(wd_heading[j+1] - (wt.heading+action-1)) + 180)

            next_Q_A = Q_table[indice_next_angle]
            Q_table[indice_angle, indice_action] += alpha * (R + discout_factor*np.max(next_Q_A) - Q_table[indice_angle, indice_action])        
    return Q_table
```

```python
# Sarsa algorithm with basic policy
def Sarsa(initial_speed = WIND_SPEED, wind_heading = WIND_HEADING, trubine_heading = TURBINE_HEADING, \
               alpha = 0.01, discout_factor = 0.95, epsion = 0.6, max_iter = MAX_ITER):
   
    len_speed = initial_speed*3
    speed_vector = np.arange(0, len_speed, 0.5)
    Q_table = np.zeros((3, 360, len(speed_vector)))

    ba = Basic_agent()

    for i in tqdm(range(max_iter)):
        wd = Wind(initial_speed, wind_heading, 60, 'OU')
        wt = Wind_turbine(trubine_heading, True)
        sm = Simu(ba, wd, wt, int(np.ceil(24*3600 / wd.step_duration)))
        sm.run_simu()
        wd_heading = sm.wd_heading_log
        wd_speed = sm.wd_speed_log
        
        for j in range(len(wd_heading)-1):
            rel_angle = wrap_to_m180_p180(wd_heading[j] - wt.heading)
            speed = wd_speed[j]        
            indice_angle = int(rel_angle) + 180
            indice_speed = int(np.round(speed/0.5))
            optimal_action = ba.policy(rel_angle)
            
            # epsilon greedy with basic policy
            if(np.random.uniform(low=0.0, high=1.0) < 1 - epsion):
                action = optimal_action
            else:
                action = int(np.random.randint(0, 3))
                
            indice_action = int(action)
            R = wt.step(speed, wd_heading[j], action)
            indice_next_angle = int(wrap_to_m180_p180(wd_heading[j+1] - (wt.heading+action-1)) + 180)
            indice_next_speed = int(np.round(wd_speed[j+1]/0.5))
            next_Q_A = [Q_table[0, indice_next_angle, indice_next_speed],\
                        Q_table[1, indice_next_angle, indice_next_speed],\
                        Q_table[2, indice_next_angle, indice_next_speed]]
            next_optimal_action = np.argmax(next_Q_A)
            if(np.random.uniform(low=0.0, high=1.0) < 1 - epsion):
                next_action = next_optimal_action
            else:
                next_action = int(np.random.randint(0, 3))
            # Compute Q(S, A) according to Sarsa algo    
            Q_table[indice_action, indice_angle, indice_speed] += alpha * (R + discout_factor*Q_table[next_action, indice_next_angle, indice_next_speed] - Q_table[indice_action, indice_angle, indice_speed])        
    return Q_table
```

```python
# Rewrite Simu_RL beacause we want go get policy according to the Q_table, speed, heading 
class Simu_RL:
    power_output_log = [] 
    action_log = []
    rel_wind_heading_log = []
    true_rel_wind_heading_log = [] 
    wd_heading_log = [] 
    step_count = 0

    def __init__(self, Q_table, wind_model=None, wind_turbine_model=None, max_steps=None):
        self.wd = Wind(10, 0, 1, 'OU') if wind_model is None else wind_model
        self.wt = Wind_turbine(0, False) if wind_turbine_model is None else wind_turbine_model
        self.max_steps = 24*3600 if max_steps is None else max_steps 
        self.power_output_log = self.max_steps * [0]
        self.action_log = self.max_steps * [1]
        self.rel_wind_heading_log = self.max_steps * [0]
        self.true_rel_wind_heading_log = self.max_steps * [0]
        self.wd_heading_log = self.max_steps * [0]
        self.wd_speed_log = self.max_steps * [0]
        
        # Add Q_table in this part
        self.Q_table = Q_table
        
    def step(self):
        self.rel_wind_heading_log[self.step_count] = wrap_to_m180_p180(self.wd.heading - self.wt.heading)
        self.true_rel_wind_heading_log[self.step_count] = wrap_to_m180_p180(self.wd.heading - self.wt.true_heading)
        self.wd_heading_log[self.step_count] = self.wd.heading
        self.wd_speed_log[self.step_count] = self.wd._speed
        
        # Get action according to Q table
        self.action_log[self.step_count] = get_policy(self.Q_table, self.rel_wind_heading_log[self.step_count], self.wd._speed)
        self.power_output_log[self.step_count] = self.wt.step(self.wd.speed, self.wd.heading, self.action_log[self.step_count])
        self.wd.step()

    def run_simu(self):
        while self.step_count < self.max_steps:
            self.step()
            self.step_count += 1

    def __str__(self):
        return str(self.__class__) + ": " + str(self.__dict__)
```

## 3. Implement a deep-Q-learning algorithm to solve the same task.

The structure of the Deep Q-learning code learned from this tutorial: https://www.youtube.com/watch?v=wc-FxNENg9U&t=96s

Since for the deep-Q-learning algorithm, the states can be continuous values, we use $s = (\theta, v)$ to represent a state, but at this time, $\theta$ and $v$ are continuous values, not the same as the previous question. 

We set state as input ad actions $a \in \{0, 1, 2 \}$ as output.

Input dimension is 2 and the output dimension is 3.

```python
class DeepQLearning(nn.Module):
    # We use two hidden layer as the neural architecture
    def __init__(self, lr=0.003, input_dims=2, fc1_dims=6, fc2_dims=6, n_actions=3):
        super(DeepQLearning, self).__init__()
        
        self.fc1 = nn.Linear(input_dims, fc1_dims)
        # initialize the weights
        self.fc1.weight.data.normal_(0, 0.1)
        
        self.fc2 = nn.Linear(fc1_dims, fc1_dims)        
        self.fc2.weight.data.normal_(0, 0.1)
        
        self.fc3 = nn.Linear(fc2_dims, n_actions)
        self.fc3.weight.data.normal_(0, 0.1)
        
        # Define the optimizer and loss function 
        self.optimizer = optim.Adam(self.parameters(), lr =lr)
        self.loss = nn.MSELoss()
        
        self.device = T.device('cuda:0' if T.cuda.is_available() else 'cpu')
        self.to(self.device)

    def forward(self, state):
        # Use relu as activation function
        x = F.relu(self.fc1(state))
        x = F.relu(self.fc2(x))
        actions = self.fc3(x)
        return actions
```

```python
class Deep_agent():
    def __init__(self, gamma=0.95, epsilon=0.8, batch_size=8, max_mem_size=100000, eps_end=0.01, eps_dec=5e-4):
        self.gamma = gamma
        self.epsilon = epsilon
        self.eps_min = eps_end
        self.eps_dec = eps_dec
        self.action_space = [i for i in range(3)]
        self.mem_size = max_mem_size
        self.batch_size = batch_size
        self.mem_cntr = 0

        self.Q_eval = DeepQLearning()
        
        # use state_memory, new_state_memory, action_memory and reward_memory to save the (state, action, reward, new_state) 
        self.state_memory = np.zeros((self.mem_size, 2), dtype=np.float32)
        self.new_state_memory = np.zeros((self.mem_size, 2), dtype=np.float32)
        self.action_memory = np.zeros(self.mem_size, dtype=np.int32)
        self.reward_memory = np.zeros(self.mem_size, dtype=np.float32)

        # Store a transition
    def store_transition(self, state, action, reward, state_):
        index = self.mem_cntr % self.mem_size
        self.state_memory[index] = state
        self.new_state_memory[index] = state_
        self.reward_memory[index] = reward
        self.action_memory[index] = action
        # Increment the counter
        self.mem_cntr += 1

    def choose_action(self, state):
        # epsilon greedy policy
        if np.random.random() > self.epsilon:
            state = T.from_numpy(state).float()
            actions = self.Q_eval.forward(state)
            action = T.argmax(actions).item()
        else:
            action = np.random.choice(self.action_space)        
        return action

    def learn(self):
        # If the transition is not enough to be used
        if self.mem_cntr < self.batch_size:
            return

        self.Q_eval.optimizer.zero_grad()

        max_mem = min(self.mem_cntr, self.mem_size)
        # Random choose a batch and get the index
        batch = np.random.choice(max_mem, self.batch_size, replace=False)
        batch_index = np.arange(self.batch_size, dtype=np.int32)

        state_batch = T.tensor(self.state_memory[batch]).to(self.Q_eval.device)
        new_state_batch = T.tensor(self.new_state_memory[batch]).to(self.Q_eval.device)
        reward_batch = T.tensor(self.reward_memory[batch]).to(self.Q_eval.device)

        action_batch = self.action_memory[batch]

        q_eval = self.Q_eval.forward(state_batch)[batch_index, action_batch]
        q_next = self.Q_eval.forward(new_state_batch)
        q_target = reward_batch + self.gamma * T.max(q_next, dim=1)[0]

        loss = self.Q_eval.loss(q_target, q_eval).to(self.Q_eval.device)
        loss.backward()
        self.Q_eval.optimizer.step()
        # change epsilon according to iteration
        self.epsilon = self.epsilon - self.eps_dec if self.epsilon > self.eps_min else self.eps_min        
```

```python
# Define an agent
agent = Deep_agent()

ba = Basic_agent()

for i in tqdm(range(200)):
    wd = Wind(WIND_SPEED, WIND_HEADING, 60, 'OU')
    wt = Wind_turbine(TURBINE_HEADING, True)
    sm = Simu(ba, wd, wt, int(np.ceil(24*3600 / wd.step_duration)))
    sm.run_simu()
    wd_heading = sm.wd_heading_log
    wd_speed = sm.wd_speed_log
    
    for j in range(len(wd_heading)-1):
        rel_angle = wrap_to_m180_p180(wd_heading[j] - wt.heading)
        speed = wd_speed[j]
        state = np.array([rel_angle/180, speed/WIND_SPEED])
        action = agent.choose_action(state)
        reward = wt.step(speed, wd_heading[j], action)
        new_state = np.array([wrap_to_m180_p180(wd_heading[j+1] - (wt.heading+action-1))/180, wd_speed[j+1]/WIND_SPEED])
        # Get a state, action, reward, new_state and store them
        agent.store_transition(state, action, reward, new_state)
        agent.learn()
```

```python
# Rewrite Simu_Deep_RL beacause we want go get policy according to the function predict
class Simu_Deep_RL:
    power_output_log = [] 			# MW
    action_log = []
    rel_wind_heading_log = []		# deg
    true_rel_wind_heading_log = [] 	# deg
    wd_heading_log = [] 			# deg
    step_count = 0

    def __init__(self, predict, wind_model=None, wind_turbine_model=None, max_steps=None):
        self.wd = Wind(10, 0, 1, 'OU') if wind_model is None else wind_model
        self.wt = Wind_turbine(0, False) if wind_turbine_model is None else wind_turbine_model
        self.max_steps = 24*3600 if max_steps is None else max_steps 
        self.power_output_log = self.max_steps * [0]
        self.action_log = self.max_steps * [1]
        self.rel_wind_heading_log = self.max_steps * [0]
        self.true_rel_wind_heading_log = self.max_steps * [0]
        self.wd_heading_log = self.max_steps * [0]
        self.wd_speed_log = self.max_steps * [0]
        self.predict = predict
        
    def step(self):
        self.rel_wind_heading_log[self.step_count] = wrap_to_m180_p180(self.wd.heading - self.wt.heading)
        self.true_rel_wind_heading_log[self.step_count] = wrap_to_m180_p180(self.wd.heading - self.wt.true_heading)
        self.wd_heading_log[self.step_count] = self.wd.heading
        self.wd_speed_log[self.step_count] = self.wd._speed
        
        # Get action according to the function
        state = T.from_numpy(np.array([self.rel_wind_heading_log[self.step_count]/180, self.wd._speed/WIND_SPEED])).float()
        actions = self.predict(state)
        # If the function only return a int type, that means it return an action
        if (type(actions) == int):
            action = actions
        # If the function return a list, we get the indice of the highest value
        else:
            action = T.argmax(actions).item()
        self.action_log[self.step_count] = action
        self.power_output_log[self.step_count] = self.wt.step(self.wd.speed, self.wd.heading, self.action_log[self.step_count])
        self.wd.step()

    def run_simu(self):
        while self.step_count < self.max_steps:
            self.step()
            self.step_count += 1

    def __str__(self):
        return str(self.__class__) + ": " + str(self.__dict__)
```

```python
# Plot the result
alphas = [0.01]
names = ["Basic_agent", "Q_learning_optimal", "Q_learning_basic", "SARSA_basic", "Q_learning_sans_speed", "Deep_q_leaning"]

for alpha_ in alphas:

    Q_learning_optimal_table = Q_learning(policy_flag = 0, alpha = alpha_)
    Q_learning_basic_table = Q_learning(policy_flag = 1, alpha = alpha_)
    Sarsa_table = Sarsa(alpha = alpha_)
    Q_learning_sans_speed_table = Q_learning_sans_speed(alpha = alpha_)
    
    sm_q_learning_optimal = Simu_RL(Q_learning_optimal_table, wd, wt, int(np.ceil(24*3600 / wd.step_duration)), )
    sm_q_learning_optimal.run_simu()

    sm_q_learning_basic = Simu_RL(Q_learning_basic_table, wd, wt, int(np.ceil(24*3600 / wd.step_duration)), )
    sm_q_learning_basic.run_simu()
    
    sm_sarsa = Simu_RL(Sarsa_table, wd, wt, int(np.ceil(24*3600 / wd.step_duration)), )
    sm_sarsa.run_simu()
    
    sm_q_learning_sans_speed = Simu_RL(Q_learning_sans_speed_table, wd, wt, int(np.ceil(24*3600 / wd.step_duration)), )
    sm_q_learning_sans_speed.run_simu()   

    sm_deep_q_learning = Simu_Deep_RL(agent.Q_eval.forward, wd, wt, int(np.ceil(24*3600 / wd.step_duration)), )
    sm_deep_q_learning.run_simu()

    power_output_q_learning_basic = sm_q_learning_basic.power_output_log 
    power_output_q_learning_optimal = sm_q_learning_optimal.power_output_log 
    power_output_sarsa = sm_sarsa.power_output_log
    power_output_q_learning_sans_speed = sm_q_learning_sans_speed.power_output_log
    
    cum_power_q_learning_optimal = np.cumsum(power_output_q_learning_optimal)    
    cum_power_q_learning_baisc = np.cumsum(power_output_q_learning_basic)
    cum_power_sarsa = np.cumsum(power_output_sarsa)
    cum_power_q_learning_sans_speed = np.cumsum(power_output_q_learning_sans_speed)
    
    power_output_deep_q_learning = sm_deep_q_learning.power_output_log
    cum_power_deep_q_learning = np.cumsum(power_output_deep_q_learning)  
    
    argv = [cum_power_ba, cum_power_q_learning_optimal, cum_power_q_learning_baisc, 
                    cum_power_deep_q_learning, cum_power_sarsa, cum_power_q_learning_sans_speed]
```

## 4. Implement a policy gradient algorithm.

This part of the code reference from this website: https://huggingface.co/blog/deep-rl-pg

We generate 100 episodes of wind; for each episode, we randomly get 20 consecutive sequences as one episode, so we have 100 episodes as training data. As the data is not big enough, we only use one hidden layer, and the size is 8.

```python
class Policy(nn.Module):

    def __init__(self, s_size=2, a_size=3, h_size=8):
        super(Policy, self).__init__()
        self.fc1 = nn.Linear(s_size, h_size)
        self.fc2 = nn.Linear(h_size, a_size)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return F.softmax(x, dim=0)
    
    # Get action and log_prob according to state
    def act(self, state):
        state = state
        probs = self.forward(state)
        m = Categorical(probs)
        action = m.sample()
        return action.item(), m.log_prob(action)
    
    # perdict the action accord to the state
    def predict(self, state):
        predict, _ = self.act(state)
        return predict
```

```python
policy = Policy()

# define a optimizer
optimizer = optim.Adam(policy.parameters(), lr=0.003)
```

```python
def reinforce(policy, optimizer, n_training_episodes = 100, gamma = 0.9):
    ba = Basic_agent()

    for i_episode in tqdm(range(1, n_training_episodes+1)):
        wd = Wind(WIND_SPEED, WIND_HEADING, 60, 'OU')
        wt = Wind_turbine(TURBINE_HEADING, True)
        sm = Simu(ba, wd, wt, int(np.ceil(24*3600 / wd.step_duration)))
        sm.run_simu()
        wd_heading = sm.wd_heading_log
        wd_speed = sm.wd_speed_log        
        
        saved_log_probs = []
        rewards = []
        random = np.random.randint(0, 1350)
        for j in range(random, random+20):
            rel_angle = wrap_to_m180_p180(wd_heading[j] - wt.heading)
            speed = wd_speed[j]
            state = np.array([rel_angle/180, speed/WIND_SPEED])
            action, log_prob = policy.act(T.from_numpy(state).float())
            saved_log_probs.append(log_prob)
            reward = wt.step(speed, wd_heading[j], action)
            new_state = np.array([wrap_to_m180_p180(wd_heading[j+1] - (wt.heading+action-1))/180, wd_speed[j+1]/WIND_SPEED])
            rewards.append(reward)
        discounts = [gamma**i for i in range(len(rewards)+1)]
        ## We calculate the return by sum(gamma[t] * reward[t]) 
        R = sum([a*b for a,b in zip(discounts, rewards)])
        
        policy_loss = []
        for log_prob in saved_log_probs:
            policy_loss.append(-log_prob * R)
        policy_loss = T.stack(policy_loss).sum()
        
        optimizer.zero_grad()
        policy_loss.backward()
        optimizer.step()        
```

```python
reinforce(policy, optimizer)
```
